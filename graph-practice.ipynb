```python
import networkx as nx

#creating an un-directed graph
sG = nx.Graph()

sG.add_node("A")
sG.add_node("B")
sG.add_node("C")
sG.add_edge("A", "B")
nx.draw(sG, with_labels = True)
```

```python
#creating a directed graph
dG = nx.DiGraph()
#adding nodes to the directed graph
dG.add_node("A")
dG.add_node("B")
dG.add_node("C")

# #add edges to the directed graph (without Weights)
# dG.add_edges_from([("A", "C"), ("A", "B"), ("B", "C")])

# add edges with weights to the directed graph
dG.add_edges_from([("A", "C", {"weight" : 2}), ("B", "C", {"weight" : 3}), ("A", "B", {"weight" : 2})])

#display the directed graph
nx.draw(dG, with_labels = True)

#print all edges of the directed graph
for i in dG.edges():
    print(i)
```

```python
#run Dijkstra shortest path algorithm on the graph
nx.shortest_path_length(dG, source="A")
```
