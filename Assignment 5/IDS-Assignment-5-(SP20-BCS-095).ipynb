```python
# Importing Libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity

import warnings
warnings.filterwarnings('ignore')
```

```python
s1 = 'sunshine state enjoy sunshine'
s2 = 'brown fox jump high, brown fox run'
s3 = 'sunshine state fox run fast'

corpus = [s1, s2, s3]
print(corpus)
```

```python
# BoW Model
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(corpus)

feats = vectorizer.get_feature_names()
vocab = vectorizer.vocabulary_
vectors = bag_of_words.toarray()

print('Features: ', feats)
print('Vocabulary: ', vocab)

print('\nVectors:')
i = 0
for i in range(0, len(vectors)):
    print(f'S{i}:', vectors[i])
    i =+ 1

bow = pd.DataFrame(bag_of_words.toarray(), columns = feats)
display(bow)
```

```python
tfidf_vectorizer = TfidfTransformer()
X = tfidf_vectorizer.fit_transform(bag_of_words)
idf = pd.DataFrame({'term':feats, 'idf':tfidf_vectorizer.idf_})
display(idf)
```

```python
tf_idf = pd.DataFrame(X.toarray() ,columns=feats)
display(tf_idf)
```

```python
# Cosine similarity b/w S1 and S3
cosine_sim = cosine_similarity(bow.iloc[[0,2]])
cosine_sim_df = pd.DataFrame(cosine_sim)
display(cosine_sim_df)
```
