```python
import folium
import pandas as pd
```

```python
punjab_map = folium.Map(location = [31.17, 72.70], zoom_start = 6.5, tiles = 'OpenStreetMap')
```

```python
#Data Source
#https://www.researchgate.net/publication/347694230_Identification_of_potential_lockdown_areas_during_COVID-19_transmission_in_Punjab_Pakistan#pf5
df = pd.read_csv('./covid-punjab-data.csv')
df
```

```python
pt = 0
for c in df['Total Cases']:
    pt = int(c) + int(pt)
```

```python
for state, lat, long, ct, agr, dr in zip(list(df['City']), list(df['Latitude']), list(df['Longitude']), list(df['Total Cases']), list(df['Avg Growth Rate']), list(df['Doubling Rate (Days)'])):
    
    folium.CircleMarker(location = [lat, long], radius = (ct/pt)*100, color = 'red', fill = False, fill_color = 'blue').add_to(punjab_map)
    
    html_format = '<strong>City:  ' + state + '</strong><br>' + 'Total Cases: ' + str(ct) + '<br>' + 'Avg Growth Rate: '+ str(agr) + '<br>' + 'Doubling Rate (Days): ' + str(dr)
    
    iframe = folium.IFrame(html_format, width = 250, height = 90)
    
    popupData = folium.Popup(iframe, max_width = 500)
    
    folium.Marker(location = [lat, long], popup = popupData).add_to(punjab_map)
    

punjab_map
```
